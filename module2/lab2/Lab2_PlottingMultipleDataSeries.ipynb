## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data_bar = data[["RetailSales", "RetailTransfers", "WarehouseSales", "ItemType"]].groupby(["ItemType"]).agg("sum")
data_bar.plot(kind = 'bar', figsize=(14,8))
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
#transfor the data
avg_sales = data.groupby(['Month'])["RetailSales", "RetailTransfers", "WarehouseSales"].mean()
data_bar.plot(kind = 'bar', figsize=(14,8))
plt.show()
#plot
avg_sales.plot(figsize=(20,10))
plt.title('', fontweight = 700)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
avg_sales = data.groupby(['Month'])["RetailSales", "RetailTransfers", "WarehouseSales"].mean()
avg_sales.plot(kind = 'line', figsize=(14,8))
plt.show()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
retail_item = pd.pivot_table(data, columns= 'RetailSales', values= 'ItemType', index=['Year','Month'], aggfunc = sum)
retail_item
```

```python
retail_item =  data.groupby(['ItemType'])["RetailSales", "ItemType"].mean()
retail_item.plot(kind = 'line', figsize=(14,8))
plt.show()

sns.lineplot(x='Year', y='RetailSales', data=data, hue='ItemType')
```

```python
data = data.astype({"Year": str}, errors='raise') 
data = data.astype({"Month": str}, errors='raise') 
data.dtypes
```

```python
data['YearMonth'] = data['Month'] + '-' + data['Year']
sns.set(rc = {'figure.figsize':(15,8)})
sns.lineplot(x='YearMonth', y='RetailSales', data=data, hue='ItemType')
```

## 6. Plot the same information as above but as a bar chart.
