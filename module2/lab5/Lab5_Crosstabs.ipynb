### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(movies.movie_title, movies.country)
```

```python
movies.groupby(['country', 'movie_title'])['movie_title'].count().unstack().fillna(0)
```

```python
pd.crosstab(movies['movie_title'].count(), movies['country'])
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
df = pd.crosstab([movies.director_name, movies.genres],[movies.country], dropna=False)
df
```

```python
movies.groupby(['director_name', 'genres', 'genres'])['movie_title'].count().unstack().fillna(0)
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(movies.genres, movies.color, values=movies.duration, aggfunc='mean', margins_name = 'Total').round(0).fillna(0)

```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
movies['Total_Likes'] = movies['actor_1_facebook_likes'] + movies['actor_2_facebook_likes'] +  movies['actor_3_facebook_likes']
```

```python
movies.groupby(["director_name"])["Total_Likes"].mean()
```

```python
x = movies.groupby(["director_name"])["Total_Likes"].mean()
x.plot.bar
```
