```python
import requests


url = "https://www.reuters.com/business/retail-consumer/americas-biggest-warehouse-is-running-out-room-its-about-get-worse-2022-08-02/"
html = requests.get(url).content
html[0:600]  #To have a little part of it 
```

```python
pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(html,"lxml")
soup
```

```python
tags = ["h1","h2","h3","h4","p"]
text = [element.text for element in soup.find_all(tags)]
text
```

```python
print('\n'.join(text))
```
